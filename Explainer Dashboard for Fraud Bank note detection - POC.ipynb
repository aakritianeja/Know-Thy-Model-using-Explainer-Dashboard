```python
import explainerdashboard

import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from explainerdashboard import RegressionExplainer, ExplainerDashboard,ClassifierExplainer

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df= pd.read_csv("BankNote_Authentication.csv")
#print the dataset
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
pred = rf.predict(X_test)
```

```python
score = accuracy_score(y_test,pred)
score
```

```python
explainer = ClassifierExplainer(rf, X_test, y_test)
#Start the Dashboard
db = ExplainerDashboard(explainer,title="Bank Note Prediction",whatif=True)
#Running the app on a local port 3050
db.run(port=3050)
```
